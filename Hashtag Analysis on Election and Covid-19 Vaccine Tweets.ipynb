```python
# EXTRACTING HASHTAGS FROM THE ELECTION TWEETS DATASET

election_data = pd.read_csv(r"C:/Users/Raheyma Arshad/Desktop/Election Tweets.csv")
del election_data['Unnamed: 0']
election_data['tweet'] = election_data['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
print('Number of election 2020 tweets:', len(election_data['tweet']))

# Defining function to return all the hashtags in a text. 
def extract_hashtags(text): 
    hashtag_list = [] 
    for word in text.split(): 
        if word[0] == '#': 
            hashtag_list.append(word[1:]) 
    return hashtag_list

election_data['Hashtags'] = election_data['tweet'].apply(lambda x: extract_hashtags(x))
election_hashtags_list = election_data['Hashtags'].tolist()
election_hashtags_flist = [item for sublist in election_hashtags_list for item in sublist]

# Getting a count of all the hashtags.
from collections import Counter
election_hashtags_counts = Counter(election_hashtags_flist)
top_election_hashtags = sorted(election_hashtags_counts, key = election_hashtags_counts.get, reverse = True)
top50_election_hashtags = top_election_hashtags[:50]
print('The top 50 election 2020 hashtags are:\n', top50_election_hashtags)

# Getting a list of all unique hashtags in the election tweets.
def unique_hashtags(list1): 
    unique_list = [] 
    for x in list1: 
        if x not in unique_list: 
            unique_list.append(x) 
    return unique_list

election_unique_hashtags = unique_hashtags(election_hashtags_flist)
```

```python
# EXTRACTING HASHTAGS FROM THE COVID-19 VACCINE TWEETS DATASET

vaccine_data = pd.read_csv(r"C:/Users/Raheyma Arshad/Desktop/Vaccine Tweets.csv")
del vaccine_data['Unnamed: 0']
vaccine_data['tweet'] = vaccine_data['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
print('Number of covid-19 vaccine tweets:', len(vaccine_data['tweet']))

# Getting all the hashtags in each tweet.
vaccine_data['Hashtags'] = vaccine_data['tweet'].apply(lambda x: extract_hashtags(x))
vaccine_hashtags_list = vaccine_data['Hashtags'].tolist()
vaccine_hashtags_flist = [item for sublist in vaccine_hashtags_list for item in sublist]

# Getting a count of all the hashtags.
vaccine_hashtags_counts = Counter(vaccine_hashtags_flist)
top_vaccine_hashtags = sorted(vaccine_hashtags_counts, key = vaccine_hashtags_counts.get, reverse = True)
top50_vaccine_hashtags = top_vaccine_hashtags[:50]
print('The top 50 covid-19 vaccine hashtags are:\n', top50_vaccine_hashtags)

# Getting a list of all unique hashtags in the vaccine tweets.
vaccine_unique_hashtags = unique_hashtags(vaccine_hashtags_flist)
```

```python
# Finding common hashtags in both election and covid-19 vaccine tweets.

election_unique_hashtags_as_set = set(election_unique_hashtags)
unique_intersection = election_unique_hashtags_as_set.intersection(vaccine_unique_hashtags)
unique_intersection_as_list = list(unique_intersection)

print('All the common hashtags in election and covid-19 tweets are:\n', unique_intersection_as_list)

# Finding common hashtags in the top 50 hashtags of both election and covid-19 vaccine tweets.
top50_election_hashtags_as_set = set(top50_election_hashtags)
top50_intersection = top50_election_hashtags_as_set.intersection(top50_vaccine_hashtags)
top50_intersection_as_list = list(top50_intersection)

print('\nAll the common hashtags in the top 50 election and covid-19 tweets hashtags are:\n', top50_intersection_as_list)
```
